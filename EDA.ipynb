# Releationship with the columns 
Step 1: Run Rate Analysis


```python
# Correlation heatmap for numeric columns related to run rate

run_rate_columns = ["Innings_run_rate", "Over_number", "Over_ball", "Innings_runs", "Over_runs"]
correlation_matrix = df[run_rate_columns].corr()

plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap - Run Rate columns")
plt.show()

# Scatter plot: Over_number vs Innings_run_rate
plt.figure(figsize=(8,9))
sns.scatterplot(data=df, x="Over_number", y="Innings_run_rate")
plt.title("Run Rate vs Over Number  ")
plt.xlabel("Over Number ")
plt.show()

# Line plot: Over_number vs Over_runs
plt.figure(figsize=(8,6))
sns.lineplot(data=df, x="Over_number", y="Over_runs")
plt.title("Runs Scored in Each Over")
plt.xlabel("Over Number")
plt.ylabel("Runs per Over")
plt.show()
```

Step 2: Wicket Fall Analysis


```python
# Count of dismissals by over

dismissals_by_over = df.groupby("Over_number")["Dismissal_is_true"].sum()

plt.figure(figsize=(8,6))
dismissals_by_over.plot(kind="bar", color="orange")
plt.title("Dismissals by Over")
plt.xlabel("Over Number")
plt.ylabel("Number of Dismissals")
plt.show()

# Scatter plot: Runs vs Wickets
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="Runs", y="Wickets", hue="Dismissal_is_true")
plt.title("Wickets vs Runs")
plt.xlabel("Total Runs")
plt.ylabel("Total Wickets")
plt.show()

# Dismissals by Bowler and Batsman
bowler_dismissals = df[df['Dismissal_is_true']].groupby("Bowler_name").size().sort_values(ascending=False).head(10)
batsman_dismissals = df[df["Dismissal_is_true"]].groupby("Batsman_name").size().sort_values(ascending=False).head(10)

# Plotting top 10 bowlers with dismissals
plt.figure(figsize=(8,9))
bowler_dismissals.plot(kind="bar", color="purple")
plt.title("Top 10 Bowlers by Dismissals")
plt.xlabel("Bowler Name")
plt.ylabel("Number of Dismissal")
plt.xticks(rotation=45)
plt.show()

# Plotting top 10 batsmen dismissed
plt.figure(figsize=(8, 6))
batsman_dismissals.plot(kind="bar", color="red")
plt.title("Top 10 Batsmen by Dismissals")
plt.xlabel("Batsman Name")
plt.ylabel("Number of Times Dismissed")
plt.xticks(rotation=45)
plt.show()
```

```python
# 1. Distribution of event-specific columns
event_columns = ["Over_no_Ball", "Over_wide", "Over_byes", "Over_leg_byes", "Dismissal_is_true"]
print(f"Event Column Value Counts")
for col in event_columns:
    print(f"\n{col}:")
    print(df[col].value_counts())

# 2. Plot distribution of event-specific columns

plt.figure(figsize=(12,6))
for i, col in enumerate(event_columns):
    plt.subplot(3,2,i+1)
    df[col].value_counts().plot(kind='bar', color='teal')
    plt.title(f"{col} Distribution")
    plt.xlabel("Event Value")
    plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
print(" Plot distribution of event-specific columns is done")

# 3. Analyze dismissals over overs

dismissals_by_over = df.groupby("Over_number")["Dismissal_is_true"].sum()
plt.figure(figsize=(8, 6))
dismissals_by_over.plot(kind="bar", color="orange")
plt.title("Dismissals by Over")
plt.xlabel("Over Number")
plt.ylabel("Number of Dismissals")
plt.show()
print("Analyze dismissals over overs iz done")


# 4. Analyze wides and no-balls by bowler

wide_balls_by_bowler = df.groupby("Bowler_name")["Over_wide"].sum().sort_values(ascending=False).head(10)
no_balls_by_bowler = df.groupby("Bowler_name")["Over_no_Ball"].sum().sort_values(ascending=False).head(10)

print("Analyze wides and no-balls by bowler is done")

# Plot top 10 bowlers contributing wides and no-balls

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
wide_balls_by_bowler.plot(kind="bar", color="purple")
plt.title("Top 10 Bowler by Wide")
plt.xlabel("Bowler Name")
plt.ylabel("Number of Wides")
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
no_balls_by_bowler.plot(kind="bar", color="red")
plt.title("Top 10 Bowlers by No-Balls")
plt.xlabel("Bowler Name")
plt.ylabel("Number of No-Balls")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("Analyze wides and no-balls by bowler is done")

# 5. Event contribution over innings
event_by_innings = df.groupby("Innings")[["Over_no_Ball", "Over_wide", "Over_byes", "Over_leg_byes"]].sum()
print("\nEvent Contributions by Innings:")
print(event_by_innings)

# Heatmap for event contributions by innings
plt.figure(figsize=(8, 6))
sns.heatmap(event_by_innings, annot=True, cmap="YlGnBu", fmt=".0f")
plt.title("Event Contributions by Innings")
plt.xlabel("Event Type")
plt.ylabel("Innings")
plt.show()
```

```python
from wordcloud import WordCloud
from collections import Counter
from sklearn.feature_extraction.text import CountVectorizer
from textblob import TextBlob
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Word Frequency Analysis
print("Generating Word Cloud...")
all_text = " ".join(df["Commentary"].astype(str))
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(all_text)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Commentary")
plt.show()

# Top 20 words
print("Analyzing Top 20 Words in Commentary...")
vectorizer = CountVectorizer(stop_words="english", max_features=20)
word_counts = vectorizer.fit_transform(df["Commentary"].astype(str))
word_frequencies = dict(zip(vectorizer.get_feature_names_out(), word_counts.toarray().sum(axis=0)))
print("Top 20 Words in Commentary:")
print(word_frequencies)

# N-Gram Analysis (Bigrams)
print("Analyzing Top 20 Bigrams in Commentary...")
vectorizer_bigrams = CountVectorizer(ngram_range=(2, 2), max_features=20, stop_words="english")
bigram_counts = vectorizer_bigrams.fit_transform(df["Commentary"].astype(str))
bigram_frequencies = dict(zip(vectorizer_bigrams.get_feature_names_out(), bigram_counts.toarray().sum(axis=0)))
print("\nTop 20 Bigrams in Commentary:")
print(bigram_frequencies)

# 2. Sentiment Analysis
print("Performing Sentiment Analysis...")
df["Commentary_sentiment"] = df["Commentary"].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
plt.figure(figsize=(8, 6))
plt.hist(df["Commentary_sentiment"], bins=20, color="skyblue", edgecolor="black")
plt.title("Sentiment Distribution in Commentary")
plt.xlabel("Sentiment Polarity")
plt.ylabel("Frequency")
plt.show()

# 3. Commentary Length Distribution
print("Analyzing Commentary Length Distribution...")
df["Commentary_length"] = df["Commentary"].apply(lambda x: len(str(x)))
plt.figure(figsize=(8, 6))
plt.hist(df["Commentary_length"], bins=20, color="green", edgecolor="black")
plt.title("Commentary Length Distribution")
plt.xlabel("Number of Characters")
plt.ylabel("Frequency")
plt.show()

# 4. Keyword Presence Analysis
print("Analyzing Keyword Presence...")
keywords = ["wicket", "six", "four", "boundary", "run"]
keyword_counts = {word: df["Commentary"].str.contains(word, case=False).sum() for word in keywords}
print("\nKeyword Frequency in Commentary:")
print(keyword_counts)

plt.figure(figsize=(8, 6))
plt.bar(keyword_counts.keys(), keyword_counts.values(), color="purple")
plt.title("Keyword Frequency in Commentary")
plt.xlabel("Keyword")
plt.ylabel("Frequency")
plt.show()

# 5. Commentary Length vs Runs/Wickets
print("Plotting Commentary Length vs Runs (Hue: Wickets)...")
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="Runs", y="Commentary_length", hue="Wickets", palette="coolwarm")
plt.title("Commentary Length vs Runs (Hue: Wickets)")
plt.xlabel("Runs")
plt.ylabel("Commentary Length")
plt.show()

# Sentiment Polarity vs Runs/Wickets
print("Plotting Sentiment Polarity vs Runs (Hue: Wickets)...")
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="Runs", y="Commentary_sentiment", hue="Wickets", palette="viridis")
plt.title("Sentiment Polarity vs Runs (Hue: Wickets)")
plt.xlabel("Runs")
plt.ylabel("Sentiment Polarity")
plt.show()

# Event Keyword Frequency
print("Plotting Frequency of Event Keywords in Commentary...")
keyword_event_counts = {word: df[df["Commentary"].str.contains(word, case=False)].shape[0] for word in keywords}
plt.figure(figsize=(8, 6))
plt.bar(keyword_event_counts.keys(), keyword_event_counts.values(), color="teal")
plt.title("Event Keyword Frequency in Commentary")
plt.xlabel("Keyword")
plt.ylabel("Frequency")
plt.show()

# Event-Specific Commentary Length
print("Analyzing Commentary Length for Different Events...")
df["Event"] = df.apply(lambda x: "Wicket" if x["Dismissal_is_true"] 
                       else "Boundary" if x["Total_Runs_on_delivery"] in [4, 6] else "Regular", axis=1)
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x="Event", y="Commentary_length")
plt.title("Commentary Length for Different Events")
plt.xlabel("Event Type")
plt.ylabel("Commentary Length")
plt.show()
```
